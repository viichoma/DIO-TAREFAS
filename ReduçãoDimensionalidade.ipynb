```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
imagem = cv2.imread('/content/Lenna_(test_image).png')
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

cv2.imwrite("cinza.jpg", imagem)

cv2_imshow(imagem)
cv2.waitKey(0)
```

```python
import cv2

# Ler a imagem em escala de cinza
imagem = cv2.imread("cinza.jpg", cv2.IMREAD_GRAYSCALE)

# Definir o valor de limiar (threshold)
limiar = 128

# Aplicar a binarização
_, imagem_bin = cv2.threshold(imagem, limiar, 255, cv2.THRESH_BINARY)

# Salvar a nova imagem
cv2.imwrite("imagem_binarizada.jpg", imagem_bin)

# Exibir a imagem original e a binarizada
cv2_imshow(imagem)
cv2_imshow(imagem_bin)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
