```python
# Importando as bibliotecas necessárias
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report

# Configurando classes e dados aleatórios
np.random.seed(42)
classes = ['Classe 0', 'Classe 1']
n_samples = 50

# Gerando rótulos verdadeiros (y_true) e previsões (y_pred) aleatórias
y_true = np.random.choice(classes, size=n_samples)
y_pred = np.random.choice(classes, size=n_samples)

# Calculando a matriz de confusão
con_mat = confusion_matrix(y_true, y_pred, labels=classes)

# Criando um DataFrame para a matriz de confusão
con_mat_df = pd.DataFrame(con_mat, index=classes, columns=classes)

# Exibindo a matriz de confusão como um heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(con_mat_df, annot=True, fmt="d", cmap=plt.cm.Blues, cbar=False)
plt.title('Matriz de Confusão 2x2')
plt.ylabel('Rótulo Verdadeiro')
plt.xlabel('Rótulo Previsto')
plt.show()

# Exibindo um relatório de classificação
print("Relatório de Classificação:")
print(classification_report(y_true, y_pred, target_names=classes))
```

```python
#Pratica de calculo de métricas Classe 0
VP = 12
VN = 18
FP = 9
FN = 11

Sensibilidade = VP / (VP + FN)
Especificidade = VN / (VN + FP)
Precisao = VP / (VP + FP)
F1 = 2 * ((Precisao * Sensibilidade) / (Precisao + Sensibilidade))
Acuracia = (VP + VN) / (VP + VN + FP + FN)

print("Sensibilidade:", Sensibilidade)
print("Especificidade:", Especificidade)
print("Precisão:", Precisao)
print("F1:", F1)
print("Acurácia:", Acuracia)
```
